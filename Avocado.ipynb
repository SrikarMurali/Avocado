```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import RFECV
from sklearn import preprocessing
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction import FeatureHasher
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.grid_search import GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
```

```python
train, test = train_test_split(data, test_size=0.3)
train["type"] = pd.get_dummies(train["type"])
test["type"] = pd.get_dummies(test["type"])
```

```python
train.head()
```

```python
all_X = train.drop(["AveragePrice", "Date", "region"], axis=1)
all_y = train["AveragePrice"]
model = LinearRegression(fit_intercept=True)
rfecv = RFECV(model, cv=10)
rfecv.fit(all_X, all_y)
best_columns = list(all_X.columns[rfecv.support_])
print("Best Columns \n"+"-"*12+"\n{}\n".format(best_columns))
```

```python
X_train, y_train = train[["Total Volume", "4046", "4225", "Total Bags"]], train["AveragePrice"]
X_test, y_test = test[["Total Volume", "4046", "4225", "Total Bags"]], test["AveragePrice"]
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
score = cross_val_score(model, X_test,y_test)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
```

```python
mae = abs(predictions - y_test)
print('Mean Absolute Error:', round(np.mean(mae), 8), 'degrees.')
```

About 31.69% error in predicting prices using linear regression.

```python
predictors = X_train.columns
coef = pd.Series(model.coef_, predictors).sort_values()
coef.plot(kind="bar", title="Model Coefficients")
```

```python
def feature_selection(df):
    #df = df.select_dtypes([np.number]).dropna(axis=1)
    all_X = df.drop(["type", "Date"], axis=1)
    all_y = df["type"]
    rfc = RandomForestClassifier(random_state=1)
    select = RFECV(rfc, cv=10)
    select.fit(all_X, all_y)
    best_columns = list(all_X.columns[select.support_])
    print("Best Columns \n"+"-"*12+"\n{}\n".format(best_columns))
    return best_columns
```

Using Lasso and Ridge yields similar results, and polynomial features does not really change it.

Now to use classification to predict the avocado type.

```python
avocado = pd.read_csv("avocado.csv")
```

A feature hasher transforms the region variable into a different type.

```python
fh = FeatureHasher(n_features=5,input_type='string')
```

```python
hashed = fh.fit_transform(avocado["region"]).toarray()
avocado = pd.concat([avocado, pd.DataFrame(hashed)], axis=1)
```

```python
avocado = avocado.drop("region", axis=1)
```

```python
X, y = avocado.drop("type", axis=1), avocado["type"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
best_features = feature_selection(avocado)
```

Several models will be used, including RandomForest and logistic regression.

```python
rfc = RandomForestClassifier(n_estimators=100)

rfc.fit(X_train[["AveragePrice", "Total Volume", "4046", "4225", "Total Bags", 0, 1, 2, 3, 4]], y_train)
```

```python
pred_rfc = rfc.predict(X_test[["AveragePrice", "Total Volume", "4046", "4225", "Total Bags", 0, 1, 2, 3, 4]])
```

```python
print(classification_report(pred_rfc, y_test))
```

```python
print(confusion_matrix(pred_rfc, y_test))
```

```python
logReg = LogisticRegression()
```

```python
logReg.fit(X_train[["AveragePrice", "Total Volume", "4046", "4225", "Total Bags", 0, 1, 2, 3, 4]], y_train)
logReg.score(X_train[["AveragePrice", "Total Volume", "4046", "4225", "Total Bags", 0, 1, 2, 3, 4]], y_train)
```

```python
pred_log = logReg.predict(X_test[["AveragePrice", "Total Volume", "4046", "4225", "Total Bags", 0, 1, 2, 3, 4]])
```

```python
print(classification_report(pred_log, y_test))
```

```python
print(confusion_matrix(pred_log, y_test))
```
